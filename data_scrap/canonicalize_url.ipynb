```python
import pandas as pd
```

```python
df = pd.read_csv("final.csv")
```

```python
df.head()
```

```python
def canonicalize_url(url):
    if not isinstance(url, str):
        return url
    return url.replace(
        "https://www.shl.com/products/",
        "https://www.shl.com/solutions/products/"
    )
```

```python
df['url'] = [canonicalize_url(u) for u in df['url']]
```
